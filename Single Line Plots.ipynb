## Matplotlib.pyplot.plt()

```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [2,3,5,2,6,6,8,3]
plt.plot(x)
```

```python
x =np.arange(10)

plt.plot(x)
```

```python
plt.plot(x,[y**2 for y in x])
```

```python
plt.plot(x,x**2)
```
