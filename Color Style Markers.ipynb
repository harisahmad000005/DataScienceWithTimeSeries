```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(5)

y= x

plt.plot(x, y+1 , 'g')
plt.plot(x, y+0.5 , 'y')
plt.plot(x, y , 'r')
plt.plot(x, y-0.2 , 'c')
plt.plot(x, y-0.4 , 'k')
plt.plot(x, y- 0.6 , 'm')
plt.plot(x, y- 0.8 , 'w')
plt.plot(x, y- 1 , 'b')

plt.show()
```

```python
#Type of markers

plt.plot(x,y,'.')

plt.plot(x,y+0.5,',')
plt.plot(x,y+1,'o')
plt.plot(x,y+2,'<')
plt.plot(x,y+3,'>')
plt.plot(x,y+4,'^')
plt.plot(x,y+5,'1')
plt.plot(x,y+6,'2')
plt.plot(x,y+7,'3')
plt.plot(x,y+8,'s')
plt.plot(x,y+9,'p')
plt.plot(x,y+10,'*')
plt.plot(x,y+11,'h')
plt.plot(x,y+12,'H')
plt.plot(x,y+13,'+')
plt.plot(x,y+14,'D')
plt.plot(x,y+15,'d')
plt.plot(x,y+16,'_')
plt.plot(x,y+17,'|')
```

```python
plt.plot(x,y, 'mo--')
plt.plot(x,y+1,'g*-.')
plt.show()
```

```python
plt.plot(x,y, color= 'g', linestyle= '--', linewidth= 1.5, 
         marker= '^' , markerfacecolor= 'b' , markeredgecolor= 'k',
         markeredgewidth= 1.5 , markersize= 5)

plt.grid(True)
plt.show()
```
