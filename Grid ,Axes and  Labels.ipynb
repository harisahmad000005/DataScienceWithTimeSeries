```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(3)
plt.grid(True)
plt.plot(x, x**2,x,x**3,x,2*x,x,2**x)
plt.show()
```

```python
x = np.arange(3)
print(plt.axis())
plt.grid(True)
plt.plot(x, x**2,x,x**3,x,2*x,x,2**x)
plt.show()
```

```python
x = np.arange(3)
print(plt.axis())

plt.plot(x, x**2,x,x**3,x,2*x,x,2**x)
plt.grid(True)

plt.axis([0,2,0,8])

plt.show()
```

```python
x = np.arange(3)
plt.plot(x, x**2,x,x**3,x,2*x,x,2**x)
plt.grid(True)

plt.xlim([0,2])
plt.ylim([0,8])
plt.show()
```

```python
#Adding Labels


x = np.arange(3)
plt.plot(x, x**2,x,x**3,x,2*x,x,2**x)
plt.grid(True)
plt.xlabel('x= np.arange(3)')
plt.xlim([0,2])
plt.ylabel('y= f(x)')
plt.ylim([0,8])
plt.show()
```

```python
#Adding Labels


x = np.arange(3)
plt.plot(x, x**2,x,x**3,x,2*x,x,2**x)
plt.grid(True)
plt.xlabel('x= np.arange(3)')
plt.xlim([0,2])
plt.ylabel('y= f(x)')
plt.ylim([0,8])
plt.title('Simple Plot Demo')
plt.show()
```

```python
#Adding Labels, legend(meaning of the figure)


x = np.arange(3)
plt.plot(x, x**2 , label='x**2')
plt.plot(x,x**3,label='x**3')
plt.plot(x,2*x,label='2**x')
plt.plot(x,2**x,label='2**x')
plt.legend()
plt.grid(True)
plt.xlabel('x= np.arange(3)')
plt.xlim([0,2])
plt.ylabel('y= f(x)')
plt.ylim([0,8])
plt.title('Simple Plot Demo')
plt.show()
```

```python


x = np.arange(3)
plt.plot(x, x**2,x,x**3,x,2*x,x,2**x)
plt.legend(['x**2','x**3','2*x','2**x'])
plt.grid(True)
plt.xlabel('x= np.arange(3)')
plt.xlim([0,2])
plt.ylabel('y= f(x)')
plt.ylim([0,8])
plt.title('Simple Plot Demo')
plt.show()
```

```python

x = np.arange(3)
plt.plot(x, x**2,x,x**3,x,2*x,x,2**x)
plt.legend(['x**2','x**3','2*x','2**x'],  loc='upper center' )
plt.grid(True)
plt.xlabel('x= np.arange(3)')
plt.xlim([0,2])
plt.ylabel('y= f(x)')
plt.ylim([0,8])
plt.title('Simple Plot Demo')
plt.show()
```

```python
#Saving the image to the disk
x = np.arange(3)
plt.plot(x, x**2,x,x**3,x,2*x,x,2**x)
plt.legend(['x**2','x**3','2*x','2**x'],  loc='upper center' )
plt.grid(True)
plt.xlabel('x= np.arange(3)')
plt.xlim([0,2])
plt.ylabel('y= f(x)')
plt.ylim([0,8])
plt.title('Simple Plot Demo')
plt.savefig('test.png')
plt.show()
```
